```python
####  merge the vector together for data_item
# Rename and replace [ ]
col = data_item.columns
col1=[]
for i in col:
    col1.append('c_'+str(i))
data_item.columns=col1
print(data_item.columns)
```

```python
    # find columns [  ] 
col_change =[]
for i in col1:
    if '[' in str(data_item.loc[0,i]) or ']' in str(data_item.loc[0,i]):
        col_change.append(i)
print(col_change) 

    # define replace function
def drop_x(x):
    if '[' in x or ']' in x:
        x=x.replace('[','')
        x=x.replace(']','')
    return x
    # replace
for i in col_change:
    data_item[i]=data_item[i].apply(lambda x: drop_x(x)) 
data_item.head()
```

```python
# merge the text_vec/img_vec to vector for data_item
text_vec_list = data_item.columns[1:129]
img_vec_list = data_item.columns[129:]
print(text_vec_list)
print(img_vec_list)
```

```python
data_item.shape
```

```python
def merge_x(x):
    vec = []
    for i in range(len(x)):
        vec.append(float(x[i]))
    return vec    
# data_item['text_vec']=data_item[text_vec_list].apply(lambda x: drop_x(x)) 
# data_item['text_vec'].head()
# data_test= data_item
data_item['text_vec']=data_item[text_vec_list].apply(lambda x: merge_x(x),axis = 1) 
print(data_item['text_vec'][0])
print(len(data_item['text_vec'][0]))
data_item['img_vec']=data_item[img_vec_list].apply(lambda x: merge_x(x),axis = 1) 
print(data_item['img_vec'][0])
print(len(data_item['img_vec'][0]))
```

```python
data_item_proc = data_item[['c_0','text_vec','img_vec']]
data_item_proc.rename(columns={'c_0':'item_id'},inplace= True)
```
