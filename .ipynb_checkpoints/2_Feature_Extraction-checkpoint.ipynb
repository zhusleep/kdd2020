```python
import pandas as pd 
import numpy as np
import datetime
from annoy import AnnoyIndex
from tqdm import tqdm
import pickle
```

```python
click_columns = ['user_id','item_id','time']
train_click_T = pd.DataFrame()
test_click_T = pd.DataFrame()
test_qtime_T = pd.DataFrame()
# 第t个阶段，读取所有文件并合并
T = 4
for i in range(T+1):
    temp_train_click = pd.read_csv('underexpose_train/underexpose_train_click-%s.csv'%(str(i)),sep=',',names = click_columns )
    temp_train_click['phrase'] = i
    temp_test_click = pd.read_csv('underexpose_test/underexpose_test_click-%s.csv'%(str(i)),sep=',',names = click_columns)
    temp_test_click['phrase'] = i
    temp_test_qtime = pd.read_csv('underexpose_test/underexpose_test_qtime-%s.csv'%(str(i)),sep=',',names = ['user_id','time'])
    temp_test_qtime['phrase'] = i
    train_click_T = train_click_T.append(temp_train_click)
    test_click_T = test_click_T.append(temp_test_click)
    c = test_qtime_T.append(temp_test_qtime)

# 拼接所有历史点击数据
data_click_T = pd.concat([train_click_T,test_click_T],axis = 0).reset_index(drop = True)
```

```python
###################  特征提取 -- item
# 商品图片/文字相似度、历史总点击次数排序、同城市/年龄/性别/点击次数
data_item = pd.read_pickle('Feature/underexpose_item_feat.pkl')
# item_id, txt_vec, img_vec
data_item.head()
```

```python
# 使用annoy算法快速计算最临近100个item
def Annoy_tree_vec(df,item_id,vec_name):
    vec_dim = len(df[vec_name][0])
    Annoy_tree = AnnoyIndex(vec_dim,metric='angular')  # 初始化新的索引树，元素向量维度为f=128
    for index,row in df.iterrows():
        Annoy_tree.add_item(row[item_id], row[vec_name])
    Annoy_tree.build(100) # 10 trees,more is better
    Annoy_tree.save('Feature/Annoy_item_%s.ann'%(str(vec_name)))  

# 对所有item构建索引树
Annoy_tree_vec(data_item,'item_id','text_vec')
Annoy_tree_vec(data_item,'item_id','img_vec')
```

```python
#Feature1: 每个item最相近的100个item,text_vec，与时间无关
Annoy_tree_text_vec = AnnoyIndex(128)
Annoy_tree_text_vec.load('Feature/Annoy_item_text_vec.ann') # super fast, will just mmap the file

sim_item_100_text={}
for index,row in tqdm(data_item.iterrows()):
    sim_item= Annoy_tree_text_vec.get_nns_by_item(row['item_id'], 100) # will find the 1000 nearest neighbors,list
    sim_item_100_text[row['item_id']]= sim_item
```

```python
#Feature2: 每个item最相近的100个item,imag_vec，与时间无关
Annoy_tree_img_vec = AnnoyIndex(128)
Annoy_tree_img_vec.load('Feature/Annoy_item_img_vec.ann') # super fast, will just mmap the file

sim_item_100_img={}
for index,row in tqdm(data_item.iterrows()):
    sim_item= Annoy_tree_img_vec.get_nns_by_item(row['item_id'], 100) # will find the 1000 nearest neighbors,list
    sim_item_100_img[row['item_id']]= sim_item

# 存储至Feature文件夹
fw = open('Feature/sim_item_100_img.pkl','wb')  
pickle.dump(sim_item_100_img,fw, protocol = pickle.HIGHEST_PROTOCOL)    
fw.close()
```

```python
#Feature3: 商品的流行度-点击次数排序 -- All phrase
data_click_item_T = data_click_T.groupby('item_id',as_index= False)['user_id'].agg({'user_dis':'unique','click_cnt':'count'})
data_click_item_T = data_click_item_T.sort_values(by = 'click_cnt',axis = 0,ascending = False)['item_id'].tolist()
#Feature3_0: 商品的流行度-点击次数排序 -- phrase 0
data_click_item_0 = data_click_T[data_click_T.phrase == 0].groupby('item_id',as_index= False)['user_id'].agg({'user_dis':'unique','click_cnt':'count'})
data_click_item_0 = data_click_item_0.sort_values(by = 'click_cnt',axis = 0,ascending = False)['item_id'].tolist()
#Feature3_1: 商品的流行度-点击次数排序 -- phrase 1
data_click_item_1 = data_click_T[data_click_T.phrase == 1].groupby('item_id',as_index= False)['user_id'].agg({'user_dis':'unique','click_cnt':'count'})
data_click_item_1 = data_click_item_1.sort_values(by = 'click_cnt',axis = 0,ascending = False)['item_id'].tolist()
#Feature3_2: 商品的流行度-点击次数排序 -- phrase 2
data_click_item_2 = data_click_T[data_click_T.phrase == 2].groupby('item_id',as_index= False)['user_id'].agg({'user_dis':'unique','click_cnt':'count'})
data_click_item_2 = data_click_item_2.sort_values(by = 'click_cnt',axis = 0,ascending = False)['item_id'].tolist()
#Feature3_3: 商品的流行度-点击次数排序 -- phrase 3
data_click_item_3 = data_click_T[data_click_T.phrase == 3].groupby('item_id',as_index= False)['user_id'].agg({'user_dis':'unique','click_cnt':'count'})
data_click_item_3 = data_click_item_3.sort_values(by = 'click_cnt',axis = 0,ascending = False)['item_id'].tolist()
#Feature3_4: 商品的流行度-点击次数排序 -- phrase 3
data_click_item_4 = data_click_T[data_click_T.phrase == 4].groupby('item_id',as_index= False)['user_id'].agg({'user_dis':'unique','click_cnt':'count'})
data_click_item_4 = data_click_item_4.sort_values(by = 'click_cnt',axis = 0,ascending = False)['item_id'].tolist()

print(len(data_click_item_T)) #71904
print(len(data_click_item_0)) #40776
print(len(data_click_item_1)) #41409
print(len(data_click_item_2)) #41031
print(len(data_click_item_3)) #42815 
```

```python
#Feature4: user_id 点击上一个商品后会点击的下一个商品，与时间无关
def next_item_by(user_list):
    '''
    针对所有用户，计算历史上某个商品点击之后下一个被点击商品的list
    返回数据格式：{'item_id':{'next_item_1':'click_count1','next_item2':'click_count2'}}
    '''
    item_seq={}
    for user_id in user_list:
        temp = data_click_T[data_click_T.user_id == user_id]
        temp_1 = temp.sort_values(by = ['phrase','time'],axis = 0,ascending = True)
        c = temp_1['item_id'].tolist()
        for index,item in enumerate(c):
            if index ==len(c)-1:break
            if item not in item_seq:
                item_seq[item] = {}
                if c[index+1] not in item_seq[item]:
                    item_seq[item][c[index+1]] = 1
                else:
                    item_seq[item][c[index+1]] += 1
            else:    
                if c[index+1] not in item_seq[item]:
                    item_seq[item][c[index+1]] = 1
                else:
                    item_seq[item][c[index+1]] += 1
    return item_seq
```

```python
# a.对所有用户
next_item_by_seq = next_item_by(data_click_T['user_id'].unique())
print(list(next_item_by_seq.items())[:4])# 查看前4个元素
```

```python
# b2.相似用户下一次点击商品
def sim_user_next_item_by(user_id,last_item):# 输入需要的user_id
    next_item=[]
    if user_id not in data_user_pro.user_id.unique():
        next_item=[]
    else:
        temp = data_user_pro[data_user_pro.user_id == user_id]['sim_user'][0] #sim_user list
        next_item_by_seq = next_item_by(temp)
    if last_item not in next_item_by_seq.keys():
        next_item =[]
    else:
        temp1 = sorted(next_item_by_seq[last_item].items(), key=lambda x: x[1], reverse=True) # 根据value的大小排序
        for i in range(len(temp1)):
            next_item.append(temp1[i][0])
    return next_item   
```

```python
#Feature5: 用户历史点击过的商品
    # 对phrase T ,要看T之前是否点击过
def data_click_his(df,T):
    data_click_t_his = df[df.phrase < T]  
    his_click = data_click_t_his.groupby('user_id',as_index= False)['item_id'].agg({'item_his':'unique'})
    return his_click

# phrase = 3 18505
user_hisitem_1 = data_click_his(data_click_T,1) # 18505
user_hisitem_1['phrase'] = 1
user_hisitem_2 = data_click_his(data_click_T,2) # 21457
user_hisitem_2['phrase'] = 2
user_hisitem_3 = data_click_his(data_click_T,3) # 23816
user_hisitem_3['phrase'] = 3
user_hisitem_4 = data_click_his(data_click_T,4) # 23816
user_hisitem_4['phrase'] = 4
print(user_hisitem_1.head())
print(user_hisitem_2.head())
print(user_hisitem_3.head())
print(user_hisitem_4.head())
```
